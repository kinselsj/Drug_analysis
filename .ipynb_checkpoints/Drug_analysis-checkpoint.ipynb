```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_data_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_drug_df = pd.merge(mouse_drug_data_df, clinical_trial_data_df, how="left", on="Mouse ID")

# Display the data table for preview
combined_drug_df.head()
```

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_drug = combined_drug_df.groupby(["Drug","Timepoint"])
# Convert to DataFrame
grouped_drug_mean = grouped_drug["Tumor Volume (mm3)"].mean()
grouped_drug_mean = grouped_drug_mean.reset_index()
# Preview DataFrame
grouped_drug_mean.head()
```

```python
grouped_drug_mean.dtypes()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

standard_error= grouped_drug["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
standard_error.reset_index().head()
# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
p_grouped_drug_mean = pd.pivot_table(grouped_drug_mean, values = "Tumor Volume (mm3)", index = "Timepoint", columns= "Drug")
# Preview that Reformatting worked
```

```python
p_grouped_drug_mean
```

```python
# Generate the Plot (with Error Bars)
drugs = grouped_drug_mean["Drug"]
time = grouped_drug_mean["Timepoint"]
fig, ax = plt.subplots()
#ax.errorbar(np.arange(0, len(standard_error), 1)+1,time, yerr=drugs)
ax.set_xlim(0, len(time) + 1)
ax.set_ylim(0, len(drugs) + 1)
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Size")
grouped_drug_mean.loc["Drug"].plot(label = "Capomulin")

plt.legend()
plt.show()
# Save the Figure
```
